# C√°lculo del Gradiente

```
#Creamos un tensor de una dimensi√≥n como variable x  = [2.0]
x = torch.tensor(2.0, requires_grad=True)

y = 9*x**4 + 2*x**3 + 3*x**2 + 6*x + 1
#derivamos y respecto a x
y.backward()

#El valor final de la derivada
x.grad
```
